#### Computational Graph

```python
# Data Flow Graph
# Hello, Tensorflow라는 하나의 노드, 엣지 없음
hello = tf.constant("Hello, Tensorflow!")

# Graph 실행을 위해서 Session() 함수로 session을 만들고
# run(노드명) 함수를 사용해서 실행한다.
sess = tf.Session()
print(sess.run(hello))

# # tensorflow 업그레이드로 session정의하고 run 해주는 과정이 생략됨
# tf.print(hello)
```

```python
# 0(더하기) - 0(데이터b)
# |
# 0(데이터a)

# 노드를 생성한다.
node1 = tf.constant(3.0, tf.float32)
node2 = tf.constant(4.0)
node3 = tf.add(node1, node2)
```

```python
# Graph의 요소로 tensor를 설명해준다.
print("node1: ", node1, "node2: ", node2)
print("node3: ", node3)

# Session 사라져서 다음과 같이 출력한다.
tf.print("node1: ", node1, "node2: ", node2)
tf.print("node3: ", node3)
```

#### Placeholder
그래프를 미리 만들어두고 실행시키는 단계에서 값들을 주고 싶다.

노드를 만들 때 placeholder라는 형태로 만들어준다.

image.png

1. 그래프를 정의한다.
    Placehoder라는 노드를 만든다.
2. Session의 형태로 run시킬 때, feed_dict를 이용해서 값을 넘겨준다.
3. 그래프가 실행되면서 필요한 내부 value를 업데이트한다.

```python
# constant가 아닌 placeholder로 노드를 만든다.
# feed_dict으로 값을 넘겨주면서 graph를 실행할 수 있다.

a = tf.placeholder(tf.float32)
b = tf.placeholder(tf.float32)
adder_node = a + b

print(sess.run(adder_node, feed_dict={a: 3,b:4.5}))
print(sess.run(adder_node, feed_dict={a: [1,3],b:[2,4]}))

# tensorflow 업데이트
# session과 placeholder를 사용하지 않고
# tf.function을 통해 함수를 정의함으로 대체
# @tf.function
# def adder(a,b):
#   return a+b


# a = tf.constant(3)
# b = tf.constant(4.5)
# tf.print(adder(a,b))

# c = tf.constant([1,3])
# d = tf.constant([2,4])
# tf.print(adder(c,d))
```
